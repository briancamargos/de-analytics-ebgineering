```python
#bibliotecas

import pandas as pd
```

```python
#Para abrir um arquivo TSV (valores separados por tabulação) em pandas, você pode usar a função read_csv() e especificar o separador como \t.
df_basics = pd.read_csv('basics.tsv', sep='\t',usecols=['tconst', 'primaryTitle']) #low_memory=False, deixa alocar o tanto de memoria necessario, retorna erro se estorar
df_basics                                                                          #usecols, seleciona para ler só as colunas selecionadas
```

```python
df_crew = pd.read_csv('crew.tsv',sep='\t')
df_crew
```

```python
merged_df = pd.merge(df_crew,df_basics, on='tconst')
merged_df
```

```python
df_title_director = pd.DataFrame({'Titulo':merged_df['primaryTitle'],'Diretor':merged_df['directors']})
df_title_director
    
```

```python
# dividir a coluna "Diretor" em uma lista de substrings com base no separador ","
df_title_director['Diretor_lista'] = df_title_director['Diretor'].str.split(',')

# selecionar o primeiro elemento de cada lista e armazenar o resultado em uma nova coluna "Diretor_primeiro"
df_title_director['Diretor_primeiro'] = df_title_director['Diretor_lista'].apply(lambda x: x[0])

# remover a coluna "Diretor_lista"
df_title_director.drop('Diretor_lista', axis=1, inplace=True)

# visualizar o resultado
print(df_title_director)
```

```python
df_title_director['Diretor'] = df_title_director['Diretor_primeiro']
df_title_director.drop('Diretor_primeiro', axis=1, inplace=True)
```

```python
print(df_title_director)
```

```python
df_names = pd.read_csv('names.tsv',sep='\t',usecols=['nconst', 'primaryName'])
rename = {'nconst':'Diretor'}
df_names.rename(columns=rename,inplace=True)
df_names
```

```python
df_title_diretor_names = pd.merge(df_names,df_title_director,on='Diretor')
df_title_diretor_names
```

```python
df_title_diretor_names['Diretor'] = df_title_diretor_names['primaryName']
df_title_diretor_names.drop('primaryName',axis=1,inplace=True)
df_title_diretor_names
```
